```python
na_count = sum(pd.isna(df).apply(any, axis=1))

print(
    f"En total hay {len(df)-1} registros.\nHay {na_count} entradas donde al menos un valor es NaN\nQuedan {len(df)-1-na_count} entradas donde ningún valor es NaN."
)
```

Las columnas que aparecen en df.describe() son las numéricas, como fue mencionado anteriormente. Podemos ver que:


*   La tarifa más cara es 512.3292
*   El promedio de edad es de 29.69911764705882
